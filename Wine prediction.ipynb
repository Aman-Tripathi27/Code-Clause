### Preprocessing of data

```python
X = wp.drop("quality", axis=1)
y = wp["quality"]
```

```python
X
```

***Making the quality column 0 or 1***

```python
Y = wp["quality"].apply(lambda y_value: 1 if y_value>=7 else 0)
Y
```

***Splitting data into Train Test Split***

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,
                                                    Y,
                                                    test_size=0.2,
                                                    random_state=3)
```

```python
Y.shape, Y_test.shape, Y_train.shape
```

***Model Training***

```python
model = RandomForestClassifier()
model.fit(X_train, Y_train)
```

***Model Evaluation***

```python
## Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
test_data_accuracy
```

```python
A = (6.6,0.16,0.4,1.5,0.044,48,143,0.9912,3.54,0.52,12.4)

#Changing it to numpy array
A_numpy_array = np.asarray(A)

#Reshaping
A_reshaped = A_numpy_array.reshape(1, -1)


prediction = model.predict(A_reshaped)
prediction

if (prediction[0]==1):
    print("Good Quality Wine")
else:
    print("Bad Quality Wine")
```
